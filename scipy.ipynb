```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Função para ajustar
def func(x, a, b):
    return a * x + b

# Geração de dados fictícios
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1])

# Ajuste da curva aos dados
params, covariance = curve_fit(func, x, y)

# Visualização dos resultados
plt.scatter(x, y, label='Dados')
plt.plot(x, func(x, *params), 'r-', label='Ajuste de Curva')
plt.legend()
plt.show()
```

```python
#2. Integração Numérica:

from scipy.integrate import quad

# Função a ser integrada
def integrand(x):
    return x**2

# Realização da integração definida
result, error = quad(integrand, 0, 1)

print(f"Resultado da integração: {result}")
print(f"Erro estimado: {error}")
#
```

```python
# 3. Análise Estatística:
from scipy.stats import norm

# Geração de dados fictícios
data = np.random.normal(size=1000)

# Cálculos estatísticos
mean, std = norm.fit(data)

# Visualização do histograma
plt.hist(data, bins=50, density=True, alpha=0.6, color='g')

# Adicionando a curva de ajuste
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std)
plt.plot(x, p, 'k', linewidth=2)

plt.show()
```
